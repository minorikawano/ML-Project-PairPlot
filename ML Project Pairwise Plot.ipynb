```python
# Create Seaborn Pairplot
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
exam = pd.read_csv("exams.csv")
exam.head()
# Load Dataset
sns.set_style('darkgrid')
sns.pairplot(exam)
#Plot Data
def size_widget(height=4, aspect=2):
    sns.pairplot(exam, height=height, aspect=aspect)
plt.show
```

Our hypthesis was that if the correlation between (math score and writing score) was positive, then a comparison between (math score and reading score) and (reading score and writing score) would also show a similar positive correlation. This pairplot agrees with our hypothesis. Data shows however, that reading score and writing score have a stronger correlation than the other two comparisons.
